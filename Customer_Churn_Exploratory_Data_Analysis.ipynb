## 1. Dataset Overview

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\user\Downloads\archive (2)\customer_churn_dataset-training-master.csv")
df.head()
```

## 2. Churn Distribution

```python
df.shape
df.info()
df.isnull().sum()
df.describe()
```

```python
df = df.drop(columns=['customer_id'], errors='ignore')
```

```python
df = df.rename(columns={'Churn': 'churn'})
df['churn'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['churn'].value_counts().plot(kind='bar')
plt.title("Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")
plt.show()
```

## 3. Churn Analysis by Contract Length

```python
import seaborn as sns

sns.countplot(data=df, x='Contract Length', hue='churn')
plt.title("Churn by Contract Length")
plt.xticks(rotation=15)
plt.show()
```

## 4. Customer Tenure and Churn


```python
sns.boxplot(data=df, x='churn', y='Tenure')
plt.title("Tenure by Churn")
plt.show()
```

```python
sns.boxplot(data=df, x='churn', y='Total Spend')
plt.title("Total Spend by Churn")
plt.show()
```

## 5. Key Insights & Business Recommendations


## Key Insights

- Customers on monthly contracts show higher churn rates
- Longer contract durations improve customer retention
- Customers with shorter tenure are more likely to churn
- Churned customers have lower total spend

## Business Recommendations

- Improve onboarding experience for new customers
- Encourage longer-term subscriptions with incentives
- Monitor early engagement to identify high-risk customers
- Focus retention strategies on new and monthly customers
